```python
import pandas as pd
import numpy as np

def create_trim_dictionary(df):
    result = {}
    trim_columns = [col for col in df.columns if 'Trim' in col]

    for year in df['Year'].dropna().unique():
        year_str = str(year)
        year_df = df[df['Year'] == year]
        result[year_str] = {}

        for model in year_df['Model_Name'].dropna().unique():
            model_str = str(model)
            model_df = year_df[year_df['Model_Name'] == model]
            trims = pd.unique(model_df[trim_columns].values.ravel())
            trims = [str(trim) for trim in trims if pd.notna(trim)]

            if not trims:
                no_trim_df = pd.DataFrame([{
                    'Year': year,
                    'Model_Name': model,
                    'Model_Number': np.nan,
                    'Trim': 'No Trims Available',
                    'Part_Number': np.nan,
                    'Description': np.nan,
                    'Accessory Description (EN)': np.nan,
                    'Accessory Description (FR)': np.nan,
                    'PartNumber': np.nan,
                    'DNET': np.nan,
                    'MSRP': np.nan,
                    'Labour Rate': np.nan,
                    'Suggested Labour Hours': np.nan,
                    'Suggested Base Retail Price w/labour': np.nan,
                    'Comments (EN)': np.nan,
                    'Comments (FR)': np.nan
                }])
                result[year_str][model_str] = {'No Trims Available': no_trim_df}
            else:
                result[year_str][model_str] = {}
                for trim in trims:
                    trim_mask = model_df[trim_columns].apply(lambda row: trim in row.values, axis=1)
                    filtered_df = model_df[trim_mask].copy()
                    filtered_df['Year'] = year
                    filtered_df['Trim'] = trim
                    filtered_df['Model_Number'] = np.nan
                    filtered_df['Part_Number'] = np.nan
                    filtered_df['Description'] = np.nan
                    selected_columns = [
                        'Year', 'Model_Name', 'Model_Number', 'Trim', 'Part_Number', 'Description',
                        'Accessory Description (EN)', 'Accessory Description (FR)', 'PartNumber',
                        'DNET', 'MSRP', 'Labour Rate', 'Suggested Labour Hours',
                        'Suggested Base Retail Price w/labour', 'Comments (EN)', 'Comments (FR)'
                    ]
                    result[year_str][model_str][trim] = filtered_df[selected_columns]

    return result
```

```python

def summarize_nested_dict(data):
    summary = []

    def traverse(d, path):
        if isinstance(d, dict):
            for key, value in d.items():
                traverse(value, path + [key])
        elif isinstance(d, pd.DataFrame):
            summary.append(path + [len(d)])

    traverse(data, [])

    # Determine the maximum depth
    max_depth = max(len(item) for item in summary)

    # Create column names
    columns = [f'level_{i+1}' for i in range(max_depth - 1)] + ['row_count']

    # Pad paths to match max depth
    padded_summary = [item[:-1] + [None]*(max_depth - len(item)) + [item[-1]] for item in summary]

    return pd.DataFrame(padded_summary, columns=columns)
```

```python

def combine_model_data(trim_dict):
    combined_dict = {}
    for model, years in trim_dict.items():
        all_dfs = []
        for year_data in years.values():
            for df in year_data.values():
                selected_columns = [
                    'Year', 'Model_Name', 'Model_Number', 'Trim',
                    'Accessory Description (EN)', 'Accessory Description (FR)', 'PartNumber',
                    'DNET', 'MSRP', 'Labour Rate', 'Suggested Labour Hours',
                    'Suggested Base Retail Price w/labour', 'Comments (EN)', 'Comments (FR)'
                ]
                df_selected = df[selected_columns].copy()
                all_dfs.append(df_selected)
        combined_dict[model] = pd.concat(all_dfs, ignore_index=True)
    return combined_dict
```

```python
def split_language_versions(combined_dict):
    fr_version = {}
    eng_version = {}
    
    # Set the language indicator
    fr_version["Lang"]="FR"
    eng_version["Lang"]="EN"
    for model, df in combined_dict.items():
        # Create FR version by excluding columns with '(EN)'
        fr_cols = [col for col in df.columns if '(EN)' not in col]
        fr_version[model] = df[fr_cols].copy()

        # Create ENG version by excluding columns with '(FR)'
        eng_cols = [col for col in df.columns if '(FR)' not in col]
        eng_version[model] = df[eng_cols].copy()

    return fr_version, eng_version
```

```python
def get_model_number(model_name, trim, year, model_db):
    # Filter the model database for the given model name and year
    filtered_db = model_db[(model_name in model_db['Description']) & ( year == model_db['Year'])]
    
    if filtered_db.empty:
        return None  # No matching model found
    
    if trim and trim != "No Trims Available":
        # Further filter by trim if provided
        filtered_db = filtered_db[trim in filtered_db['Trim']]
    
    if not filtered_db.empty:
        return filtered_db.iloc[0]['Model_Number']  # Return the first matching model number
    
    return None  # No matching trim found
```

```python
def export_dict_to_excel(data_dict, output_directory):
    # Extract prefix from the first key
    keys = list(data_dict.keys())
    prefix = data_dict["Lang"]

    # Prepare the output filename with timestamp
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    filename = f"Hyundai_ACCY_{prefix}_{timestamp}.xlsx"
    output_path = os.path.join(output_directory, filename)

    # Create Excel writer
    with pd.ExcelWriter(output_path, engine='openpyxl') as writer:
        for model in keys[1:]:
            
            df = data_dict[model]
            if prefix == "EN":
                final_column_names = { 
                        'Model_Number':'Model', 
                        'Accessory Description (EN)':'Description', 
                        'PartNumber':'Part', 
                        'MSRP':'Cost',
                        'Suggested Labour Hours':'Hours',
                        'Suggested Base Retail Price w/labour':"Price", 
                        'Comments (EN)':"Comments"}
                print("Done Here ")
            elif prefix == "FR":
                final_column_names = { 
                        'Model_Number':'Model', 
                        'Accessory Description (FR)':'Description', 
                        'PartNumber':'Part', 
                        'MSRP':'Cost',
                        'Suggested Labour Hours':'Hours',
                        'Suggested Base Retail Price w/labour':"Price", 
                        'Comments (FR)':"Comments"}
                
            df.rename(columns=final_column_names, inplace=True)

            df = df[["Year", 
                    "Model", 
                    "Description", 
                    "Part",  
                    "Cost", 
                    "Hours",
                    "Price",
                    "Comments",
                        "Trim", 
                        'Model_Name']]
            sheet_name = f"{model}-{prefix}"
            
            # before wriing to the shet
            df.to_excel(writer, sheet_name=sheet_name, index=False)

    print(f"All worksheets have been successfully written to '{output_directory}'.")
```
